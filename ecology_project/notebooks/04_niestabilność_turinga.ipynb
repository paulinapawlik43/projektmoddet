# Analiza niestabilności Turinga 

Zbadamy teraz, czy stabilny stan stacjonarny (wyznaczony w poprzednim etapie) może ulec destabilizacji pod wpływem dyfuzji, prowadząc do powstania wzorów przestrzennych (Niestabilność Turinga).

### 1. Wybór punktu równowagi.

Rozważamy bezwymiarowy układ równań:
$$
\begin{cases}
f(u, v) = a - u - uv^2 = 0 \\
g(u, v) = uv^2 - mv = 0
\end{cases}
$$

Z poprzedniej analizy wiemy, że dla standardowych parametrów ($a > 2m$) istnieją dwa stany stacjonarne roślinności. Interesuje nas ten stabilny lokalnie, o wyższej biomasie ($v^* > 1$).

$$v^* = \frac{a + \sqrt{a^2 - 4m^2}}{2m}, \quad u^* = \frac{m}{v^*}$$

Będziemy traktować $u^*$ i $v^*$ jako znane stałe.

### 2. Linearyzacja i macierz Jacobiego

Linearyzujemy część reakcyjną wokół punktu $(u^*, v^*)$. Macierz Jacobiego $\mathbf{J}$ ma postać:

$$
\mathbf{J} = \begin{pmatrix}
\frac{\partial f}{\partial u} & \frac{\partial f}{\partial v} \\
\frac{\partial g}{\partial u} & \frac{\partial g}{\partial v}
\end{pmatrix}_{(u^*, v^*)} = 
\begin{pmatrix}
-(1 + (v^*)^2) & -2u^*v^* \\
(v^*)^2 & 2u^*v^* - m
\end{pmatrix}
$$

Ponieważ w stanie równowagi zachodzi $u^*v^* = m$, możemy uprościć wyrazy macierzy:

$$
\mathbf{J} = \begin{pmatrix}
-(1 + (v^*)^2) & -2m \\
(v^*)^2 & m
\end{pmatrix}
$$

### 3. Warunki stabilności lokalnej (bez dyfuzji)

Aby stan był stabilny bez wpływu przestrzennego, zaburzenia muszą wygasać. Wymaga to spełnienia dwóch warunków dla macierzy $\mathbf{J}$:
1.  **Ślad macierzy:** $\text{Tr}(\mathbf{J}) = J_{11} + J_{22} < 0$
2.  **Wyznacznik:** $\det(\mathbf{J}) = J_{11}J_{22} - J_{12}J_{21} > 0$

W naszym przypadku (dla bujnej roślinności $v^*$) te warunki są zazwyczaj spełnione (co pokazaliśmy numerycznie w notatniku 02).

### 4. Uwzględnienie dyfuzji (metoda Fouriera)

Rozważamy teraz pełny układ z dyfuzją. Niech $\mathbf{w}(t, x)$ będzie małym zaburzeniem od stanu równowagi. Zlinearyzowane równanie ewolucji zaburzenia przyjmuje postać:

$$\frac{\partial \mathbf{w}}{\partial t} = \mathbf{J}\mathbf{w} + \mathbf{D} \nabla^2 \mathbf{w}$$

gdzie $\mathbf{D} = \text{diag}(d_1, d_2)$ to macierz współczynników dyfuzji.
Szukamy rozwiązań w postaci falowej (rozdzielenie zmiennych w przestrzeni Fouriera):
$$\mathbf{w}(t, x) \propto e^{\lambda t} e^{ikx}$$

Wstawiając to do równania, operator laplasjanu $\nabla^2$ działając na $e^{ikx}$ daje czynnik $-k^2$. Otrzymujemy problem wartości własnych dla nowej macierzy $\mathbf{M}(k)$:

$$\lambda \mathbf{w} = (\mathbf{J} - k^2 \mathbf{D}) \mathbf{w}$$

Relacja dyspersji $\lambda(k)$ (zależność szybkości wzrostu zaburzenia od liczby falowej) jest określona przez wartości własne macierzy:

$$\mathbf{M}(k) = \begin{pmatrix} J_{11} - d_1 k^2 & J_{12} \\ J_{21} & J_{22} - d_2 k^2 \end{pmatrix}$$

### 5. Warunek Powstawania Wzorów

Aby układ stał się niestabilny przestrzennie, musimy znaleźć takie $k \neq 0$, dla którego część rzeczywista wartości własnej macierzy $\mathbf{M}(k)$ jest dodatnia ($\text{Re}(\lambda) > 0$).

Wartości własne $\lambda$ są pierwiastkami równania charakterystycznego:
$$
\det(\mathbf{M}(k) - \lambda \mathbf{I}) = 0
$$
Co prowadzi do równania kwadratowego:
$$
\lambda^2 - \text{Tr}_k \lambda + \Delta_k = 0
$$

Gdzie współczynniki zależą od liczby falowej $k$:
* **Ślad:** $\text{Tr}_k = \text{Tr}(\mathbf{J}) - k^2(d_1 + d_2)$
* **Wyznacznik:** $\Delta_k = \det(\mathbf{J} - k^2 \mathbf{D})$

#### Analiza Znaku $\lambda$
Aby pojawiła się niestabilność ($\text{Re}(\lambda) > 0$), musi zostać naruszony jeden z warunków stabilności.
1.  Wiemy, że bez dyfuzji układ jest stabilny, więc $\text{Tr}(\mathbf{J}) < 0$. Ponieważ $k^2, d_1, d_2 > 0$, to ślad z dyfuzją $\text{Tr}_k$ jest zawsze ujemny.
2.  Skoro ślad jest ujemny, jedynym sposobem na uzyskanie dodatniej wartości własnej jest sprawienie, aby wyznacznik był ujemny:
    $$
    \Delta_k < 0
    $$

#### Warunki na parametry
Rozpiszmy wyznacznik $\Delta_k$ jako funkcję zmiennej $k^2$:
$$
\Delta_k = (J_{11} - d_1 k^2)(J_{22} - d_2 k^2) - J_{12}J_{21}
$$
Po uporządkowaniu otrzymujemy funkcję kwadratową od $k^2$:
$$
h(k^2) = (d_1 d_2) (k^2)^2 - (d_1 J_{22} + d_2 J_{11}) k^2 + \det(\mathbf{J})
$$

Wykresem tej funkcji jest parabola skierowana ramionami w górę. Aby przyjmowała wartości ujemne (warunek niestabilności) dla pewnych $k \neq 0$, muszą zostać spełnione dwa warunki konieczne:

**Różnica dyfuzji**

Minimum paraboli musi znajdować się po prawej stronie osi rzędnych. Oznacza to, że współczynnik przy wyrazie liniowym musi być ujemny (wyrażenie w nawiasie musi być dodatnie):
$$
d_1 J_{22} + d_2 J_{11} > 0
$$
W modelu Klausmeiera $J_{11} < 0$ (zanik wody) i $J_{22} > 0$ (wzrost biomasy). Aby suma była dodatnia, konieczne jest, aby:
$$
d_1 \gg d_2
$$
Oznacza to, że inhibitor (woda) musi dyfundować znacznie szybciej niż aktywator (biomasa).

**Istnienie miejsc zerowych**

Minimum funkcji musi leżeć poniżej zera ($\Delta_{min} < 0$), co oznacza, że wyróżnik tej funkcji kwadratowej musi być dodatni:
$$
(d_1 J_{22} + d_2 J_{11})^2 > 4 d_1 d_2 \det(\mathbf{J})
$$

Gdy te nierówności są spełnione, istnieje pewien zakres liczb falowych $k \in (k_{min}, k_{max})$, dla których $\text{Re}(\lambda(k)) > 0$. Zaburzenia o tych częstościach będą rosły wykładniczo, tworząc stabilne wzory przestrzenne.

```python
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
sys.path.append(os.path.abspath(os.path.join('..')))

from pipeline.niestabilnosc import calculate_dispersion_relation

A = 1.0
M = 0.45
D1 = 1
D2 = 0.01

k_valss, re_lam = calculate_dispersion_relation(A, M, D1, D2)

plt.figure(figsize=(10, 6))
plt.plot(k_vals, re_lam, 'b-', linewidth=2, label=r'Re($\lambda_{max}$)')
plt.axhline(0, color='k', linewidth=1, linestyle='--')

plt.fill_between(k_vals, 0, re_lam, where=(re_lam > 0), 
                 color='red', alpha=0.2, label='Obszar niestabilności (wzory)')

if np.max(re_lam) > 0:
    max_idx = np.argmax(re_lam)
    k_dom = k_vals[max_idx]
    val_dom = re_lam[max_idx]
    plt.plot(k_dom, val_dom, 'ro')
    plt.text(k_dom, val_dom + 0.02, f'max niestabilność k={k_dom:.2f}', ha='center')

plt.title(f'Relacja dyspersji (a={A}) - warunek powstawania wzorów')
plt.xlabel('Liczba falowa k')
plt.ylabel(r'Współczynnik wzrostu Re($\lambda$)')
plt.ylim(-0.5, 0.5)
plt.xlim(0, 7)
plt.legend()
plt.grid(True, alpha=0.5)

plt.savefig("../relacja_dyspersji_turing.png", dpi=300, bbox_inches='tight')
plt.show()
```

##### Analiza Wyników: Relacja Dyspersji i Niestabilność Turinga

Wykres przedstawia zależność rzeczywistej części największej wartości własnej $\text{Re}(\lambda)$ od liczby falowej $k$ dla modelu Klausmeiera ($a=1.0$) przy silnej dyfuzji inhibitora ($d_1=1.0, d_2=0.01$).

Wnioski:

**1. Stabilność stanu jednorodnego ($k=0$)**

Dla liczby falowej $k=0$, wartość $\text{Re}(\lambda)$ jest ujemna.
* Układ spełnia pierwszy warunek Turinga. Bez udziału dyfuzji, jednorodny stan równowagi jest stabilny. Oznacza to, że system nie dąży do "eksplozji" biomasy ani całkowitego wymarcia, lecz do ustalonego poziomu równowagi.

**2. Pasmo niestabilności**

Dla zakresu liczb falowych $k \in (1.4, 4.8)$, krzywa przecina oś zerową, przyjmując wartości dodatnie ($\text{Re}(\lambda) > 0$).
* W tym zakresie dyfuzja wody destabilizuje równowagę. Zaburzenia o tych częstotliwościach są wzmacniane (rosną wykładniczo w czasie), co prowadzi do spontanicznego łamania symetrii i formowania się wzorów przestrzennych.

**3. Najszybciej rosnąca fala**

Krzywa osiąga wyraźne maksimum dla liczby falowej **$k_{max} \approx 3.38$**. Chociaż niestabilne jest całe pasmo fal, to zaburzenie o tej konkretnej częstotliwości rośnie najszybciej i zdominuje strukturę ekosystemu w początkowej fazie ewolucji.
